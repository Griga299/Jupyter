# График синуса matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_chart():
    
    # Данные для графика
    t = np.arange(0.0, 2.0, 0.01)
    s = 1 + np.sin(2 * np.pi * t)

    fig, ax = plt.subplots()
    ax.plot(t, s)

    ax.set(xlabel='time (s)', ylabel='voltage (mV)',
           title='График синуса matplotlib')
    ax.grid()

    plt.show()
```

```python
plot_chart()
```

# Брилианты

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("P1_diamonds.csv", header = 0)
data = data.drop("Unnamed: 0", axis =1)
data.head()
```

# Графики

```python
plt.hist(data["price"], bins = 150)

# Добавление заголовка
plt.title("Частотный график: Цены на брилианты\n", fontdict={'fontsize':20, 'fontweight' : 20, 'color' : 'Green'})

plt.xlabel("Цена", fontdict={'fontsize':12, 'fontweight' : 20, 'color' : 'Brown'})
plt.ylabel("Количество", fontdict={'fontsize':12, 'fontweight' : 20, 'color' : 'Brown'})

plt.show()
```

```python
groupdata = data.groupby("cut").count()
groupdata

x_vals = []
y_vals = []
for i in [0, 1, 4, 3, 2]:
    x_vals.append(groupdata.index[i])
    y_vals.append(groupdata.iloc[i,0])

plt.bar(x_vals, y_vals)

plt.title("Частотный график: Цены на брилианты\n", fontdict={'fontsize':20, 'fontweight' : 20, 'color' : 'Green'})

plt.xlabel("Огранка", fontdict={'fontsize':12, 'fontweight' : 20, 'color' : 'Brown'})
plt.ylabel("Количество", fontdict={'fontsize':12, 'fontweight' : 20, 'color' : 'Brown'})

plt.show()
```
